```python
from IPython.display import Latex

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
```

```python
plt.rcParams['figure.figsize'] = (8,6)
ref = pd.read_csv('data/reflection.csv')
ref['l'] = 2*ref.l
delta_V = 20e-3
delta_t = 5e-9

f = lambda t, l_0, v: l_0 + v*t
popt, pcov = curve_fit(f, ref.t, ref.l)
perr = np.sqrt(np.diag(pcov))
ref['dB/m'] = 10*np.log10(ref.V_back / ref.V_foward) / ref.l
ref['delta_dB/m'] = 10*delta_V*np.sqrt(1/ref.V_foward**2 + 1/ref.V_back**2) / ref.l

plt.errorbar(ref.t, ref.l, xerr=delta_t, fmt='.', ms=10)
plt.plot(ref.t, f(ref.t.values, *popt), color='r')
plt.xlabel('t [s]', fontsize=16)
plt.ylabel('$\ell$ [m]', fontsize=16)
plt.title('$\mathbf{\ell}$(t)', fontsize=20, weight='bold')
plt.grid()
plt.show()

plt.errorbar(ref.l, ref['dB/m'], yerr=ref['delta_dB/m'], fmt='.', ms=10)
plt.xlabel('$\ell$ [m]', fontsize=16)
plt.ylabel('Attenuazione [dB/m]', fontsize=16)
plt.title('Attenuazione in funzione di $\mathbf{\ell}$', fontsize=20, weight='bold')
plt.grid()
plt.show()

display(Latex(f'v = ({popt[1]*1e-6:.0f} $\pm$ {perr[1]*1e-6:.0f}) Mm/s'))
display(Latex(f'v_0 = ({popt[0]:.1f} $\pm$ {perr[0]:.1f}) Mm/s'))
```
