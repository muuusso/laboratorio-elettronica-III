```python
from IPython.display import Latex

from matplotlib.backends.backend_pdf import PdfPages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit

pp = PdfPages('multipage.pdf')
```

```python
plt.rcParams['figure.figsize'] = (8,6)
ref = pd.read_csv('data/reflection.csv')
ref['l'] = 2*ref.l
delta_V = 3e-3
delta_t = 5e-9

f = lambda t, l_0, v: l_0 + v*t
popt_v, pcov = curve_fit(f, ref.t, ref.l)
perr_v = np.sqrt(np.diag(pcov))
ref['dB'] = 10*np.log10(ref.V_back / ref.V_foward)
ref['delta_dB'] = 10*delta_V*np.sqrt(1/ref.V_foward**2 + 1/ref.V_back**2)

plt.errorbar(ref.t*1e9, ref.l, xerr=delta_t*1e9, fmt='.', ms=10)
plt.plot(ref.t*1e9, f(ref.t.values, *popt_v), color='r')
plt.xlabel('t [ns]', fontsize=16)
plt.ylabel('$\ell$ [m]', fontsize=16)
plt.title('$\mathbf{\ell}$(t)', fontsize=20, weight='bold')
plt.grid()
pp.savefig()
plt.show()


f = lambda t, dB_0, dBm: dB_0 + dBm*t
popt, pcov = curve_fit(f, ref.l[1:], ref.dB[1:], sigma=ref.delta_dB[1:])
perr = np.sqrt(np.diag(pcov))

plt.errorbar(ref.l, ref.dB, yerr=ref.delta_dB, fmt='.', ms=10)
plt.plot(ref.l, f(ref.l.values, *popt), color='r')
plt.xlabel('$\ell$ [m]', fontsize=16)
plt.ylabel('G [dB]', fontsize=16)
plt.title('$\mathbf{G(\ell)}$', fontsize=20, weight='bold')
plt.grid()
pp.savefig()
plt.show()

display(Latex(f'$v = ({popt_v[1]*1e-8:.2f} \pm {perr_v[1]*1e-8:.2f})\cdot 10^8$ m/s'))
display(Latex(f'$\ell_0 = ({popt_v[0]:.1f} \pm {perr_v[0]:.1f})$ m/s'))

c = 299792458
n = c / popt_v[1]
delta_n = c * 1 / popt_v[1]**2 *perr_v[1]

display(Latex(f'$n = {n:.2f} \pm {delta_n:.2f}$'))

display(Latex(f'$\\alpha = ({popt[1]:.4f} \pm {perr[1]:.4f})$ dB/m'))
```

```python
pp.close()
```
