```python
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib.pyplot as plt
import numpy as np

pp = PdfPages('multipage.pdf')
```

```python
plt.rcParams['figure.figsize'] = (8,6)
R_1 = 1e6
R_2 = 1000

y_inf = np.linspace(-3, 3)
x_inf = np.sqrt(9 - y_inf**2)

y_C = np.linspace(-0.1,0.1)
x_C = np.sqrt(0.1**2 - y_C**2)

plt.plot((0,0), (0.1,3))
plt.plot((0,0), (-0.1,-3))
plt.plot(x_C, y_C)
plt.plot(x_inf, y_inf)
plt.grid()
plt.gca().set_aspect("equal")
plt.xlabel('Re(s)', fontsize=14)
plt.ylabel('Img(s)', fontsize=14)
pp.savefig()
plt.show()

y_minus = np.linspace(-1e6, -1e-6)
y_plus = np.linspace(1e-6, 1e6)

plt.plot(-y_plus**2 * (R_1 + R_2)**2 / (2*R_2)**2, y_plus)
plt.plot(-y_minus**2 * (R_1 + R_2)**2 / (2*R_2)**2, y_minus)
plt.plot(-y_plus**3 *(((R_1 + R_2)**2 / (2*R_2)**2 * 1e12) + 1e17) / 1e18 + 1e17, y_plus)
plt.plot(-y_plus**3 *(((R_1 + R_2)**2 / (2*R_2)**2 * 1e12) + 1e17) / 1e18 + 1e17, -y_plus, c='g')
plt.scatter([0],[0], c='r')
plt.xlabel('Re', fontsize=14)
plt.ylabel('Img', fontsize=14)
plt.grid()
pp.savefig()
plt.show()
```

```python
pp.close()
```
