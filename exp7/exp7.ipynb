```python
from IPython.display import Latex

from matplotlib.backends.backend_pdf import PdfPages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.optimize import curve_fit
```

```python
plt.rcParams['figure.figsize'] = (8,6)
R = 100

df = pd.read_csv('data/scope_0.csv')
df['I'] = df.V_in / R * 1e3

I = lambda V, R_0, alpha: 1e3 * V / R_0 / (1 + alpha * V**2)
popt, pcov = curve_fit(I, df.V_out, df.I)

plt.plot(df.V_out, df.I, label='dati')
plt.plot(df.V_out, I(df.V_out, *popt), label='fit')
plt.grid()
plt.show()

print(*popt)
```
